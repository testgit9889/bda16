5. Draw a pie as in picture, with the same:<br>
values
order of values
Legend,
start position,
explode,
colors,
Legend with title and position
figure size


```python
# add data values to lables
y = np.array([10, 25, 35, 40, 60, 65])
l = ["Paper", "Plastic", "Glass", "Cartons", "Wood", "Steel"]

# convert np array to list
y_list = y.tolist()


# converting the elements of the list to str and adding " " (space)
i = 0
while i < len(y_list):
    y_list[i] = " " + str(y_list[i])
    i += 1


# concationate the values of the 2 lists
l_new = [i + j for i, j in zip(l, y_list)] 


e = [0,0,0,0,0.2,0.4]
c = ["#f6ccff", "#ea80ff", "#d91aff", "#bf00e6", "#7f0099", "#550066"]

plt.figure(figsize=(30,12))

plt.pie(y, labels = l_new, textprops=dict(color="w"), startangle = 90, explode = e, colors = c)

plt.legend(title = "Recycling", loc="upper left")

plt.show()
```

We have 2 ways to do so, the first one is to copy the code and add "plt.savefig()" method just before "plt.show()". <br>
It has to be before or else we'll get a blank picture, it's because after "plt.show()" is called, a new figure is created.

```python
# add data values to lables
y = np.array([10, 25, 35, 40, 60, 65])
l = ["Paper", "Plastic", "Glass", "Cartons", "Wood", "Steel"]

# convert np array to list
y_list = y.tolist()


# converting the elements of the list to str and adding " " (space)
i = 0
while i < len(y_list):
    y_list[i] = " " + str(y_list[i])
    i += 1


# concationate the values of the 2 lists
l_new = [i + j for i, j in zip(l, y_list)] 


e = [0,0,0,0,0.2,0.4]
c = ["#f6ccff", "#ea80ff", "#d91aff", "#bf00e6", "#7f0099", "#550066"]

plt.figure(figsize=(30,12))

plt.pie(y, labels = l_new, textprops=dict(color="w"), startangle = 90, explode = e, colors = c)

plt.legend(title = "Recycling", loc="upper left")









plt.savefig('pie.png') # --> we've added "plt.savefig()" before "plt.show()"




plt.show()



```

If we wanred to save picture after "plt.show()" is called, we can Save the figure before we show() it by calling "plt.gcf()" for "get current figure". <br>
Then we can call "plt.savefig()" on this Figure object at any time.

First we'll create the figure as before, notuce we don't use "plt.savefig()", this time instead we'll use "plt.gcf" and save it to "fig" 

```python
# add data values to lables
y = np.array([10, 25, 35, 40, 60, 65])
l = ["Paper", "Plastic", "Glass", "Cartons", "Wood", "Steel"]

# convert np array to list
y_list = y.tolist()


# converting the elements of the list to str and adding " " (space)
i = 0
while i < len(y_list):
    y_list[i] = " " + str(y_list[i])
    i += 1


# concationate the values of the 2 lists
l_new = [i + j for i, j in zip(l, y_list)] 


e = [0,0,0,0,0.2,0.4]
c = ["#f6ccff", "#ea80ff", "#d91aff", "#bf00e6", "#7f0099", "#550066"]

plt.figure(figsize=(30,12))

plt.pie(y, labels = l_new, textprops=dict(color="w"), startangle = 90, explode = e, colors = c)

plt.legend(title = "Recycling", loc="upper left")






fig = plt.gcf() # --> we've added "plt.gcf" before "plt.show()" and saved the picture to fig







plt.show()
```
